```python
from kogito.core.processors.models.swem import SWEMClassifier

model = SWEMClassifier.from_pretrained("hmodels/swem")
```

```python
from relation_modeling_utils import load_fdata
from kogito.core.processors.models.swem import SWEMHeadDataset
import numpy as np
import pytorch_lightning as pl
from torch.utils.data import DataLoader

VOCAB = np.load("data/vocab_glove_100d.npy", allow_pickle=True).item()

test_df = load_fdata(f"data/atomic_ood2/n1/test_n1.csv")
test_data = SWEMHeadDataset(test_df, vocab=VOCAB)
test_dataloader = DataLoader(test_data, batch_size=128, shuffle=True)
trainer = pl.Trainer()
trainer.test(model, dataloaders=test_dataloader)
```

```python
predict_data = SWEMHeadDataset(test_df.drop(columns=['label']), vocab=VOCAB)
predict_dataloader = DataLoader(predict_data, batch_size=128, shuffle=True)
trainer = pl.Trainer()
preds = trainer.predict(model, dataloaders=predict_dataloader)
```
